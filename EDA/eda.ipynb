## Unificando Dataframes

```python
# contar quantidade de registros por dataframe

len(df_contracts)
```

```python
len(df_customers)
```

```python
len(df_services)
```

```python

df_customers.columns = ['IDCliente','Genero','Acimade60','Parceiro','Dependente']
df_customers.head()
```

```python
df_contracts.rename(columns={'customerID':'IDCliente'},inplace=True)
df_contracts.head()
```

```python
df_services.rename(columns={'customerID':'IDCliente'},inplace=True)
df_services.head()
```

```python
df_temp = df_customers.merge(df_services,on=['IDCliente'])
```

```python
df_temp
```

```python
df_churn_temp=df_temp.merge(df_contracts, left_on='IDCliente',right_on='IDCliente')
df_churn_temp
```

```python
df_churn= df_customers.merge(df_services,on=['IDCliente']).merge(df_contracts,on=['IDCliente'])
```

```python
df_churn
```
